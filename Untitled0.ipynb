```python
# prompt: a line plot

from matplotlib import dates
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import yfinance as yf

# Download data
raw = yf.download('SPY AAPL', start='2010-01-01', end='2019-12-31')

# Create the plot
plt.figure(figsize=(12, 10))  # Adjust figure size as needed
for col in ['Adj Close']: # Select the column you want to plot
    plt.plot(raw.index, raw[col]['SPY'], label='SPY') # Plot SPY data
    plt.plot(raw.index, raw[col]['AAPL'], label='AAPL') # Plot AAPL data


plt.xlabel('Date')
plt.ylabel('Adjusted Close Price')
plt.title('SPY and AAPL Adjusted Close Prices (2010-2019)')
plt.legend()
plt.grid(True)
plt.show()
```
